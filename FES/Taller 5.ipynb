# Modelo de Ising

```python
import numpy as np
import matplotlib.pyplot as plt

m = np.linspace(-1,1, 250)

tan_h1 = np.tanh(3*m+1)

tan_h2 = np.tanh(3*m-1)

tan_h3 = np.tanh(3*m)

plt.figure(dpi=150)
plt.plot(m, tan_h1, color='#3591BF', label=r'$\beta = 1$, $h = 1$, $J = 1$, $z = 3$')
plt.plot(m, tan_h2, color='#9B36BB', label=r'$\beta = 1$, $h = -1$, $J = 1$, $z = 3$')
plt.plot(m, tan_h3, color='#44BF35', label=r'$\beta = 1$, $h = 0$, $J = 1$, $z = 3$')
plt.plot(m,m, color='black', linestyle= '--', label='Referencia')
plt.legend(fontsize=9)



plt.xlabel('Magnetización (m)')
plt.ylabel('Función de la magnetización (m)')

plt.show()
```

# Teoría de Ginsburg-Landau

```python
#Importamos las librerias

import numpy as np
from scipy import integrate
from scipy.optimize import minimize_scalar

# Definimos el interior de la integral F(r,s)

def F(r, s):
    S_1 = 1/2 * (s ** 4/((r ** 2 + s ** 2) ** (3)))
    S_2 = 1/2 * (1/(r ** 2 + s ** 2))
    S_3 = 1/4 * (1 - r ** 2/(r ** 2 + s ** 2)) ** 2
    Sum = (S_1 + S_2 + S_3) * r
    return Sum
    
# Definición de fución de extremizacion de F

def extremizar(s):
    Sum, _ = integrate.quad(F, 0, np.inf, args=(s,))
    return  Sum

optimizar = minimize_scalar(extremizar)

print("Valor óptimo del parámetro: ", optimizar.x)
print("Valor extremizado de la función: ", optimizar.fun)
```
