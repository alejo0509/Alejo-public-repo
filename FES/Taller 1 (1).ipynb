```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def factorial(num):
    if num == 0 or num == 1:
        return 1
    return num * factorial(num - 1)

def binomial_pmf(k, n, p):
    # Función de probabilidad de la distribución binomial
    return (p ** k) * ((1 - p) ** (n - k)) * (factorial(n) / (factorial(k) * factorial(n - k)))


def plot_binomial_and_normal_distribution(n, p):
    x = list(range(0, n + 1))
    binomial_probabilities = [binomial_pmf(k, n, p) for k in x]
    
    mu = n * p
    sigma = np.sqrt(n * p * (1 - p))
    normal_distribution = stats.norm.pdf(x, loc=mu, scale=sigma)
    
    
    
    plt.bar(x, binomial_probabilities, label='Binomial')
    plt.plot(x, normal_distribution, color='orange', label='Gaussiana')

    plt.xlabel('Lanzamientos impares')
    plt.ylabel('Probabilidad / Densidad')
    plt.title('Distribuciones Binomial y Gaussiana')
    plt.xticks(x)
    plt.legend()
    plt.show()

n = 12
p = 1/4
plot_binomial_and_normal_distribution(n, p)
```
