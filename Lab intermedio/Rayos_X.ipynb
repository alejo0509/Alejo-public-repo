```python
#Función para la ley de Bragg

import math

def Ley_de_Bragg(angulos, d, desfase,n):
    """
    Calcula la longitud de onda de una lista de ángulos en grados utilizando la ley de Bragg.

    Args:
        angulos (list[float]): Lista de ángulos en grados.
        d (float): Parámetro de red.
        desfase (float) : ángulo de desfase en grados determinado en la calibración.

    Return:
        list[float]: Lista de longitudes de onda correspondientes a los ángulos dados.
    """
    longitud_de_onda = []
    desfase_en_radianes = math.radians(desfase)

    for angulo in angulos:
        # Convertir el ángulo de grados a radianes
        angulo_en_radianes = math.radians(angulo)
        
        # Calcular la longitud de onda para cada ángulo
        lambda_ = (2 * d * math.sin(angulo_en_radianes+desfase_en_radianes)) / n
        longitud_de_onda.append(lambda_)

    return longitud_de_onda

# Se pasa de ángulos a longitudes de onda

d = 2.014 #Å
n = 1  # Orden de difracción (generalmente 1 para primer orden)
#Note que cuando se pone n=2, las lineas Kα y Kβ se asocian a los segundos picos encontrados tal como se esperaría.

desfase = 1.8 #grados
L = Ley_de_Bragg(A,d,desfase,n) #Å
```

```python
#Función de propagación de error

import sympy as sp

def propagacion_de_error(expr, variables, errores):
    """
    Calcula el error propagado para una expresión dada utilizando el método de derivadas parciales.

    Parámetros:
        expr (str): La expresión matemática en forma de cadena.
        variables (list): Lista de símbolos de las variables en la expresión.
        errores (list): Lista de los errores asociados a cada variable.

    Retorna:
        float: Expreseion del del error propagado.
    """
    # Convertir las cadenas a símbolos de sympy
    symbols = sp.symbols(variables)
    expr = sp.sympify(expr)
    
    # Calcular el error propagado utilizando derivadas parciales
    derivadas = [sp.diff(expr, var) for var in symbols]
    error_propagado = sp.sqrt(sum([(derivada * error)**2 for derivada, error in zip(derivadas, errores)]))
    
    return error_propagado.evalf()
```

```python
#Referencia:
#https://scipython.com/book/chapter-8-scipy/examples/the-voigt-profile/

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define la función Lorentziana
def lorentzian(x, amplitude, center, width):
    return amplitude / (1 + ((x - center) / width)**2)

# Define la función Voigt
from scipy.special import wofz
def voigt(x, amplitude, center, sigma, gamma):
    return amplitude * np.real(wofz(((x - center) + 1j * gamma) / (sigma * np.sqrt(2)))) / (sigma * np.sqrt(2 * np.pi))


# Ajuste de Lorentziana
initial_guess_lorentzian = [max(I_pico1 ), np.mean(L_pico1), 1.0]  # Valores iniciales para ajuste
params_lorentzian, covariance_lorentzian = curve_fit(lorentzian, L_pico1, I_pico1 , p0=initial_guess_lorentzian, maxfev=5000)

# Ajuste de perfil Voigt
initial_guess_voigt = [max(I_pico1), np.mean(L_pico1), 0.1 , 0.12]  # Valores iniciales para ajuste
params_voigt, covariance_voigt = curve_fit(voigt, L_pico1, I_pico1 , p0=initial_guess_voigt, maxfev=5000)

# Genera curvas ajustadas
L_pico1_fit = np.linspace(min(L_pico1), max(L_pico1), 1000)
I_pico1_fit_lorentzian = lorentzian(L_pico1_fit, *params_lorentzian)
I_pico1_fit_voigt = voigt(L_pico1_fit, *params_voigt)

# Calcula los valores predichos por las funciones ajustadas
I_pico1_pred_lorentzian = lorentzian(L_pico1, *params_lorentzian)
I_pico1_pred_voigt = voigt(L_pico1, *params_voigt)

# Calcula los residuos
residuos_lorentzian = I_pico1 - I_pico1_pred_lorentzian
residuos_voigt = I_pico1 - I_pico1_pred_voigt

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))


# Grafica los datos y las curvas ajustadas
ax1.scatter(L_pico1, I_pico1, label='Datos',color='black')
ax1.plot(L_pico1_fit, I_pico1_fit_lorentzian, 'r-', label='Ajuste Lorentziana')
ax1.plot(L_pico1_fit, I_pico1_fit_voigt, 'g-', label='Ajuste Voigt')
#ax1.set_xlabel('longitud de onda (Å)')
ax1.set_ylabel('I (Imp/s)')
ax1.set_title('Pico Kβ')
ax1.legend()

# Grafica los residuos
ax2.scatter(L_pico1, residuos_lorentzian, label='Residuos Lorentziana', color='red')
ax2.scatter(L_pico1, residuos_voigt, label='Residuos Voigt', color='green')
ax2.axhline(0, color='black', linestyle='--', linewidth=0.5)
ax2.set_xlabel('longitud de onda (Å)')
ax2.set_ylabel('Residuos')
ax2.legend()

plt.subplots_adjust(hspace=0.0)

plt.show()


# Imprime los parámetros ajustados
print(f"Parámetros ajustados para Lorentziana: Amplitud = {params_lorentzian[0]}, Centro = {params_lorentzian[1]}, Ancho = {params_lorentzian[2]}")
print(f"Parámetros ajustados para Voigt: Amplitud = {max(I_pico1_fit_voigt)}, Centro = {params_voigt[1]}, Gamma = {-params_voigt[3]}")
```

En este caso se evidendencia que ambos ajustes se adaptan de manera muy similar. Los residuales toman un patron para nada aleatorio, no obstante no está asociado con los picos genereados por las distribuciones.

```python
# Ajuste de Lorentziana
initial_guess_lorentzian = [max(I_pico2 ), np.mean(L_pico2), 1.0]  # Valores iniciales para ajuste
params_lorentzian, covariance_lorentzian = curve_fit(lorentzian, L_pico2, I_pico2 , p0=initial_guess_lorentzian, maxfev=5000)

# Ajuste de perfil Voigt
initial_guess_voigt = [max(I_pico2), np.mean(L_pico2), 0.1 , 0.12]  # Valores iniciales para ajuste
params_voigt, covariance_voigt = curve_fit(voigt, L_pico2, I_pico2 , p0=initial_guess_voigt, maxfev=5000)

# Genera curvas ajustadas
L_pico2_fit = np.linspace(min(L_pico2), max(L_pico2), 1000)
I_pico2_fit_lorentzian = lorentzian(L_pico2_fit, *params_lorentzian)
I_pico2_fit_voigt = voigt(L_pico2_fit, *params_voigt)

# Calcula los valores predichos por las funciones ajustadas
I_pico2_pred_lorentzian = lorentzian(L_pico2, *params_lorentzian)
I_pico2_pred_voigt = voigt(L_pico2, *params_voigt)

# Calcula los residuos
residuos_lorentzian = I_pico2 - I_pico2_pred_lorentzian
residuos_voigt = I_pico2 - I_pico2_pred_voigt

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))


# Grafica los datos y las curvas ajustadas
ax1.scatter(L_pico2, I_pico2, label='Datos',color='black')
ax1.plot(L_pico2_fit, I_pico2_fit_lorentzian, 'r-', label='Ajuste Lorentziana')
ax1.plot(L_pico2_fit, I_pico2_fit_voigt, 'g-', label='Ajuste Voigt')
#ax1.set_xlabel('longitud de onda (Å)')
ax1.set_ylabel('I (Imp/s)')
ax1.set_title('Pico Kα')
ax1.legend()

# Grafica los residuos
ax2.scatter(L_pico2, residuos_lorentzian, label='Residuos Lorentziana', color='red')
ax2.scatter(L_pico2, residuos_voigt, label='Residuos Voigt', color='green')
ax2.axhline(0, color='black', linestyle='--', linewidth=0.5)
ax2.set_xlabel('longitud de onda (Å)')
ax2.set_ylabel('Residuos')
ax2.legend()

plt.subplots_adjust(hspace=0.0)

plt.show()

# Imprime los parámetros ajustados
print(f"Parámetros ajustados para Lorentziana: Amplitud = {params_lorentzian[0]}, Centro = {params_lorentzian[1]}, Ancho = {-params_lorentzian[2]}")
print(f"Parámetros ajustados para Voigt: Amplitud = {max(I_pico2_fit_voigt)}, Centro = {params_voigt[1]}, Gamma = {params_voigt[3]}")
```

En este caso, se evidencia un mejor ajuste en el perfil de Voigt. Además de ello, los residuales toman un patron aleatorio en la zona de los picos de la distribución lo que garantiza un buen ajuste.

# 2) Absorción de rayos X

La formula de absorción debido a rayos X se define como:

$$ I = I_0 e^{-\mu d} $$

De forma logaritmica:

$$ \tilde{I} = \ln\left( \frac{I}{I_0} \right)  $$

$$ \tilde{I} = -\mu d $$

Donde $I$ es la intensidad, $I_0$ es la intensidad inicial, $\mu$ es el coeficiente de absorción y $d$ es el espesor del material. La propagación del error al volver la ecuación logaritmica es:

$$ \Delta  \tilde{I} = \Delta I \sqrt{\frac{1}{I_0^2} + \frac{1}{I^2}} $$

$$ \Delta I = \Delta I_0 = 1.0 \; Imp/s $$

De este modo, para hallar $\mu$:

$$ \mu = - \frac{\tilde{I}}{d} $$

$$ \Delta \mu = \frac{1}{d} \sqrt{ \tilde{I}^2 \left ( \frac{\Delta d}{d} \right )^{2} + \Delta \tilde{I}^2 } $$

Ahora, $\mu$ tambien es una funcion que depende del número atómico ($Z$) y de la longitud de onda ($\lambda$):

$$ \frac{\mu}{\rho} = (k Z^3) \; \lambda^3 $$

Donde $\rho$ es la densidad del marial y $k$ es una constante. Ahora, como ya habiamos visto, para $\lambda$:

$$ \lambda = \frac{2d\sin(\theta + \delta)}{n} $$

$$ \Delta \lambda = \left | \frac{2d\cos(\theta + \delta) \Delta \theta }{n} \right |  $$

$$ \Delta \theta = 0.1 ° $$

Entonces, si:

$$ y = \frac{\mu}{\rho} $$

$$ m = (k Z^3) $$

$$ x = \lambda^3 $$

Entonces note que se puede hacer una regreción por minimos cuadrados y se tiene la siguiente propagación para el error:

$$ \Delta y = \left | \frac{\Delta \mu}{\rho} \right | $$

$$ \Delta x = \left | 3 \lambda^2 \Delta \lambda  \right |$$

$$ \Delta m = \frac{1}{x} \sqrt{ y^2 \left ( \frac{\Delta x}{x} \right )^{2} + \Delta y^2 } $$

Note que $m$ será una medida de la dependencia del amortiguamineto $\mu$ del número atómico. Entre mayor sea $m$ más grande será el número atómico del materíal (siempre y cuando se compare a la misma longitud de onda).

Recuerde que la formula de propagación de error es:

$$ \nu = \nu(x_1,\cdots ,x_N) $$

$$ \Delta \nu = \sqrt{\sum_{i=1}^{N} \left ( \frac{\partial \nu }{\partial x_i} \Delta  x_i \right )^{2} }  $$
