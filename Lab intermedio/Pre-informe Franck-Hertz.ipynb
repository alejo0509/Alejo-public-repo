```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
K_B = 1.38e-23  # Constante de Boltzmann en J/K
sigma = 2.9e-19  # cross-sectión en m^2

# Definir la función en micrometros
def calcular_MFP(T):
    num = K_B * T * 1e+6
    deno = np.sqrt(2) * sigma * 8.7 * (10**(9-(3110/T)))
    return num / deno

# Rango de temperaturas de 300K a 500K
temperaturas = np.linspace(300, 500, 1000)

# Calcular los valores de mean free path para cada temperatura
valores_MFP = calcular_MFP(temperaturas)

# Graficar la función principal
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(temperaturas, valores_MFP, color='b', label=r'$\bar{\ell} = \frac{K_B T}{\sqrt{2}\sigma} 8,7\times 10^{ \frac{3110}{T} - 9}$')
ax1.set_xlabel('Temperatura (K)', fontsize=12)
ax1.set_ylabel(r'mean free path, $\bar{\ell}$', fontsize=12)
ax1.tick_params(axis='both', labelsize=12)
ax1.legend()

# Crear el gráfico de zoom en la esquina superior derecha
ax2 = plt.axes([0.63, 0.5, 0.25, 0.25])  # [left, bottom, width, height]
ax2.plot(temperaturas, valores_MFP, color='b')
ax2.set_xlim(400, 500)
ax2.set_ylim(0, 100)
ax2.set_xticks([413, 433, 453, 473, 493])
ax2.set_yticks([0,  20, 40, 60, 80, 100])
ax2.tick_params(axis='both', labelsize=10)

plt.show()
```
