```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

def regresion_lineal(X, Y):
    # Calcular la pendiente (m) y la ordenada al origen (b) de la regresión lineal
    m, b = np.polyfit(X, Y, 1)
    
    # Calcular el error de la pendiente
    error_pendiente = np.std(Y - (m * X + b))
    
    # Calcular el coeficiente de determinación (R²)
    Y_pred = m * X + b
    ss_tot = np.sum((Y - np.mean(Y))**2)
    ss_res = np.sum((Y - Y_pred)**2)
    r2 = 1 - (ss_res / ss_tot)
    
    # Crear un arreglo de valores ajustados usando la ecuación de la regresión
    Y_pred = m * X + b
    
    # Crear la gráfica de la regresión lineal
    plt.figure(figsize=(8, 6))
    plt.scatter(X, Y, label='Datos reales')
    plt.plot(X, Y_pred, color='red', label=f'Regresión Lineal: y = {m:.2f}x + {b:.2f}\nR² = {r2:.2f}')
    plt.xlabel(r'$U_1$ (V)', fontsize=24)
    plt.ylabel('Orden del mínimo', fontsize=24)
    plt.title(r'Cálculo de $\frac{\mathrm{d \Delta E} }{\mathrm{d} n}$', fontsize=24)
    plt.legend(fontsize=15)
    plt.rc('font', family='Times New Roman')  # Configurar la fuente en Times New Roman
    plt.rc('xtick', labelsize=15)  # Tamaño de los números en el eje x
    plt.rc('ytick', labelsize=15)  # Tamaño de los números en el eje y
    plt.savefig('regresion_lineal.png')  # Guardar la gráfica de la regresión lineal
    plt.show()
    
    # Crear la gráfica de los residuales
    plt.figure(figsize=(8, 6))
    residuals = Y - Y_pred
    plt.scatter(X, residuals, color='green', label='Residuales')
    plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)  # Línea horizontal en y=0
    plt.xlabel(r'$U_1$ (V)', fontsize=24)
    plt.ylabel('Residuales', fontsize=24)
    plt.title('Gráfico de Residuales', fontsize=24)
    plt.legend(fontsize=15)
    plt.rc('font', family='Times New Roman')  # Configurar la fuente en Times New Roman
    plt.rc('xtick', labelsize=15)  # Tamaño de los números en el eje x
    plt.rc('ytick', labelsize=15)  # Tamaño de los números en el eje y
    plt.savefig('residuales.png')  # Guardar la gráfica de los residuales
    plt.show()
    
    return m, error_pendiente, r2, b


# Función para calcular E_alpha
def calcular_E_alpha(l, L, E):
    return E(0.5)/(1 + (2 * 0.5 + 1) * l / L)


def error_E_alpha(E, l, L, delta_l):
    term1 = (2 * 0.5 + 1) / (1 + (2 * 0.5 + 1) * (l / L))**2
    delta_E_alpha = abs(term1) * delta_l * E(0.5)
    return delta_E_alpha
```

```python
# Ejemplo de uso
X = np.array([3, 4, 5, 6, 7, 8])
Y = np.array([4.76, 4.81, 5.01, 5.1, 5, 5.06])

pendiente, error, r2, b = regresion_lineal(X, Y)

a = 12/(2*4.67)

l = a * pendiente
error_l = abs(a) * error


print(f'Pendiente de la regresión: {pendiente}')
print(f'Error de la pendiente: {error}')
print(f'Coeficiente de determinación (R²): {r2}')
print(f'l = {l:.2f}')
print(f'Error de l: {error_l:.2f}')

L = 12

E = lambda x: int(pendiente)*x + int(b)


# Calcular E_alpha y su error
E_alpha= calcular_E_alpha(l, L, E)
error_E_alpha_a = error_E_alpha(E, l, L, error_l)

print(f'E_alpha = {E_alpha:.2f}')
print(f'Error de E_alpha = {error_E_alpha_a:.2f}')
```
