# Oscilaciones de Rabi

## Datos

```python
import numpy as np

# Datos para Rb 85
periodo_rb85 = np.array([0.23, 0.35, 0.4, 0.5, 0.6, 0.95])
vpp_rb85 = np.array([4.2, 2.7, 2.4, 1.9, 1.66, 1.11])
inv_vpp_rb85 = np.array([0.238095238, 0.37037037, 0.416666667, 0.526315789, 0.602409639, 0.900900901])

# Datos para Rb 87
periodo_rb87 = np.array([0.18, 0.21, 0.25, 0.3, 0.4, 0.6])
vpp_rb87 = np.array([3.31, 2.7, 2.5, 2.0, 1.53, 0.9])
inv_vpp_rb87 = np.array([0.302114804, 0.37037037, 0.4, 0.5, 0.653594771, 1.111111111])
```

## Rubidio 85

```python
x_data = inv_vpp_rb85
y_data = periodo_rb85
```

```python
#Ajuste sobre los datos de la toma 1

import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Función lineal
def func(x, m, b):
    return m * x + b

# Realizar la regresión no lineal
initial_guess = [1,0]  # Valores iniciales para los parámetros
params, covariance = curve_fit(func, x_data, y_data, p0=initial_guess)

# Calcular las incertidumbres a partir de la diagonal de la matriz de covarianza
param_uncertainties = np.sqrt(np.diag(covariance))

# Imprimir los parámetros ajustados y sus incertidumbres

for i, param in enumerate(params):
    param_name = f'Param_{i+1}'
    uncertainty = param_uncertainties[i]
    print(f'{param_name}: {param:.4f} ± {uncertainty:.4f}')

# Calcular los residuales
residuals = y_data - func(x_data, *params)

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Graficar los datos originales y la curva ajustada

x_prima = np.linspace(0.24,0.9,500) #Se genera un dominio para la regreción

ax1.scatter(x_data, y_data, label="Datos originales", color='black')
ax1.plot(x_prima, func(x_prima, *params), label="Curva ajustada", color='black',linestyle='dotted')
ax1.set_xlabel('1/Vpp(1/V)')
ax1.set_ylabel('T (μs)')
ax1.legend()
ax1.set_title('Oscilaciones de Rabi Rb 85 ')

# Graficar los residuales en la parte inferior sin espacio entre los subplots
ax2.plot(x_data, residuals, 'o', color='green')
ax2.axhline(0, color='black', linestyle='--')
ax2.set_xlabel('1/Vpp (1/V)')
ax2.set_ylabel('Residuales [T (μs)] ')


# Ajustar los márgenes del segundo subplot
plt.subplots_adjust(hspace=0)

# Mostrar la figura
plt.show()
```
