## Parte a)

```python
import math

def calcular_serie(n):
    suma = 0.0
    for k in range(n):
        termino = (960 / (math.pi ** 6)) * (1 / (2 * k + 1) ** 6)
        suma += termino
    return suma

n = 100000  # n tendindo a infinito
resultado = calcular_serie(n)
print(f"Resultado de la serie con {n} términos: {round(resultado,2)}")
```

## Parte  b)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Definimos la función Ψ teniendo en cuenta que para los casos paso n = 2k esta 0. Así, tomamos n -> 2k+1 pues 
#no tiene sentido tomas los 0 de Ψ. Los indicies de k van desde 0 a infinito.

def psi_cuadrado(k, a, x):
    cte = (1920 / (a * (math.pi)**6))
    frac = (1 / ((2 * k + 1)))**6
    sin = (math.sin(((2 * k + 1) * math.pi * x) / a))**2

    return cte * frac * sin

# Dominio de las posiciones (x) dentro del poso [0,a]
a = 1
x = np.linspace(0, a, 1000)
```

```python
k = 0  # modo normal n = 2k+1

# Calcular psi_cuadrado para los valores de x y guardarlos en una lista
psi_valores = [psi_cuadrado(k, a, val_x) for val_x in x]

# Crear la gráfica
plt.plot(x, psi_valores)
plt.title(f'|Ψ|^2 vs x (n={2*k+1})')
plt.xlabel(r'$x$ (unidades arbitrarias de longitud)')
plt.ylabel(r'$\left | \Psi \right |^2$ (unidades arbitrarias de intensidad)')

# Mostrar la gráfica
plt.show()
```
