```python
import numpy as np
import matplotlib.pyplot as plt

h = 6.626e-34 # J s
c = 3.000e+8 # m/s
k = 1.381e-23 # J/K

def ley_planck(L, T):
    return (8 * np.pi * h / (L*(1e-9))**3) * (1 / (np.exp(h * c / (k * T * L * (1e-9))) - 1)) 

# Parámetros

T1 = 3000  # T1 (K)
T2 = 5000  # T2 (K)
T3 = 8000  # T3 (K)
T4 = 10000  # T4 (K)
L_range = np.linspace(0.1, 2300 , 500)  # Rango de λ en nm

# Densidad espectral de energía por unidad de frecuencia

distribucion1 = ley_planck(L_range, T1)
distribucion2 = ley_planck(L_range, T2)
distribucion3 = ley_planck(L_range, T3)
distribucion4 = ley_planck(L_range, T4)

# Graficar la densidad espectral de energía

plt.plot(L_range, distribucion1, label=f'T = {T1} K')
plt.plot(L_range, distribucion2, label=f'T = {T2} K')
plt.plot(L_range, distribucion3, label=f'T = {T3} K')
plt.plot(L_range, distribucion4, label=f'T = {T4} K')


plt.xlabel('Longigitud de onda λ(nm)')
plt.ylabel('Densidad de Energía por Unidad de Frecuencia')
plt.title('Ley de Planck del cuerpo negro')
plt.legend()
plt.show()
```
