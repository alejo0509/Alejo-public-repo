```python
#importar las librerias que se van a emplear para el desarrollo del código
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

#constantes a utilizar 
h = 6.626e-34
c = 3.000e+8
k = 1.381e-23
pi = np.pi

#se define la función de Planck
def f(wave, T):
    num = (8*pi*h)/(wave**3)
    x = (h*c/(wave*k*T))
    denom = ((np.exp(x))-1)
    funcion = (num/denom)
    return funcion

#rangos de longitudes de onda
wave = np.arange(1e-9,3e-6,1e-9)

T1 = 10000
T2 = 8000
T3 = 5000
T4 = 3000

#evaluando en las funciones
distribucion10000 = f(wave,T1)
maximo10000 = np.argmax(distribucion10000)
distribucion8000 = f(wave,T2)
maximo8000 = np.argmax(distribucion8000)
distribucion5000 = f(wave,T3)
maximo5000 = np.argmax(distribucion5000)
distribucion3000 = f(wave,T4)
maximo3000 = np.argmax(distribucion3000)

#maximos para cada T
punto10000 = [maximo10000]
punto8000 = [maximo8000]
punto5000 = [maximo5000]
punto3000 = [maximo3000]

#Grafica para las T
plt.plot(wave*1e9,distribucion10000*1e15, marker = "o", markevery = punto10000)
plt.plot(wave*1e9,distribucion8000*1e15, marker = "o", markevery = punto8000)
plt.plot(wave*1e9,distribucion5000*1e15, marker = "o", markevery = punto5000)
plt.plot(wave*1e9,distribucion3000*1e15, marker = "o", markevery = punto3000)

plt.xlabel('Longitud de onda λ (nm)')
plt.ylabel('Densidad de Energía/Frecuencia $(\mu J/m^3)/GHz$')
plt.title('Ley de Planck del cuerpo negro')

plt.show()

#calcular máximos
max_distribucion1 = np.max(distribucion10000)*1e15
max_distribucion2 = np.max(distribucion8000)*1e15
max_distribucion3 = np.max(distribucion5000)*1e15
max_distribucion4 = np.max(distribucion3000)*1e15

# Encontrar los máximos y las longitudes de onda correspondientes
max_L1 = wave[maximo10000]*1e9
max_L2 = wave[maximo8000]*1e9
max_L3 = wave[maximo5000]*1e9
max_L4 = wave[maximo3000]*1e9


print(f"Para T igual a {T1} K:")
print(f"El valor maximo de la distibucion se alcanza en {max_distribucion1} (μJ/m^3)/GHz")
print(f"El valor maximo de longitud de onda se alcanza en {max_L1} nm y hace parte del infrarrojo")

print(" ")

print(f"Para T igual a {T2} K:")
print(f"El valor maximo de la distibucion se alcanza en {max_distribucion2} (μJ/m^3)/GHz")
print(f"El valor maximo de longitud de onda se alcanza en {max_L2} nm y hace parte del infrarrojo")

print(" ")

print(f"Para T igual a {T3} K:")
print(f"El valor maximo de la distibucion se alcanza en {max_distribucion3} (μJ/m^3)/GHz")
print(f"El valor maximo de longitud de onda se alcanza en {max_L3} nm y hace parte del espectro visible (Naraja)")

print(" ")

print(f"Para T igual a {T4} K:")
print(f"El valor maximo de la distibucion se alcanza en {max_distribucion4} (μJ/m^3)/GHz")
print(f"El valor maximo de longitud de onda se alcanza en {max_L4} nm y hace parte del espectro visible (Verde)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

#constantes a utilizar 
h = 6.626e-34
c = 3.000e+8
k = 1.381e-23
pi = np.pi

#definimos funcion planck
def planck (frec,T):
    part1 = (8*pi*h)/(c**3)
    y = (h*frec)/(k*T)
    part2 = (np.exp(y))-1
    part3 = (frec**3)/part2
    distribucion_planck = part1*part3
    return distribucion_planck

#definimos funcion Rayleigh
def rayleigh (frec,T):
    part4 = (8*pi*((frec)**2))/(c**3)
    part5 = k*T
    distri_ray = part4*part5
    return distri_ray

#rango de frecuencias
frec = np.arange (1e+13,150e+13,1e+13)  

T1 = 8000

distri_frec = planck(frec,T1)
distri_ray = rayleigh(frec, T1)

#gráfica para las gráficas
plt.plot(frec*1e-13,distri_frec*1e+15, label="Planck") 
plt.plot(frec*1e-13,distri_ray*1e+15, label="Rayleigh-Jeans")
plt.ylim(0,4.8)
plt.legend(loc="upper right")

plt.xlabel('Frecuencia ν (THz)')
plt.ylabel('Densidad de Energía/Frecuencia $(\mu J/m^3)/GHz$')

plt.show()

#correlacion de las gráficas
correlacion_pr = np.corrcoef(distri_frec,distri_ray)[0,1]

print("La correlación entre la gráfica de Planck y Rayleigh es: "+ str(correlacion_pr)) 

print ("NOTA: El coeficiente de correlación entre las gráficas de Planck y Rayleigh-jeans tienen una correlacion negativa es decir que hay una tendencia a evolucionar en direcciones opuestas.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

#constantes a utilizar 
h = 6.626e-34
c = 3.000e+8
k = 1.381e-23
pi = np.pi

#definimos Planck
def planck (frec,T):
    part1 = (8*pi*h)/(c**3)
    y = (h*frec)/(k*T)
    part2 = (np.exp(y))-1
    part3 = (frec**3)/part2
    distribucion_planck = part1*part3
    return distribucion_planck

#definimos Wien
def ley_wien(frec,T):
    p1 = (8*pi*h*(frec**3))/(c**3)
    x = (h*frec)/(k*T)
    p2 = np.exp (-x)
    wien = p1*p2
    return wien

#Rango de frecuencias
frec = np.arange (1e+13,150e+13,1e+13)  
T1 = 8000

distri_frec = planck(frec,T1)
distri_wien = ley_wien(frec, T1)

#Graficamos
plt.plot(frec*1e-13,distri_frec*1e+15, label="Planck") 
plt.plot(frec*1e-13,distri_wien*1e+15, label="Wien")
plt.ylim(0,4.8)
plt.legend(loc="upper right")

plt.xlabel('Frecuencia ν (THz)')
plt.ylabel('Densidad de Energía/Frecuencia $(\mu J/m^3)/GHz$')

plt.show()

#Correlación entre ambas gráficas
correlacion_pw = np.corrcoef(distri_frec,distri_wien)[0,1]

print("La correlación entre la gráfica de Planck y Wein es: "+ str(correlacion_pw)) 
print ("NOTA: El coeficiente de correlación entre las gráficas de Planck y Wien tienen una correlacion positiva es decir que evolucionan en el mismo sentido.")
```
